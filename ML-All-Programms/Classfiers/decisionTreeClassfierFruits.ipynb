```python
import pandas as pd
import numpy as np
data = pd.read_csv(r"E:\PES\SEM2\Refer\ML\ML_Datasets-20230613T052137Z-001\ML_Datasets\fruits.csv")
```

```python
x = data[['Weight','Size']]
y = data['Class']
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x,y)
```

```python
ypred = clf.predict(x)
```

```python
ypred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y,ypred)
```

```python
ypred
```

```python
from sklearn.tree import plot_tree
plot_tree(clf,feature_names=["Weight","Size"],class_names = ["orange","apple"],filled=True)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = GridSearchCV(DecisionTreeClassifier(),
                     param_grid={"criterion":["gini","engtropy"],"splitter":['best','ranodm']},cv = 5,verbose = True)
```

```python
best_model = model.fit(x,y)
```

```python
best_model.best_estimator_
from sklearn.tree import plot_tree
plot_tree(best_model.best_estimator_,feature_names=["Weight","size"],class_names=["Orange","apple"],filled=True)
```
