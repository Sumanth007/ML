```python
import matplotlib.pyplot as plt
```

```python
x = [1.25,1.0,3.0,4.0,4.0]
```

```python
y = [5,1,1,2,34]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
z = [1,2,3]
m = [1,2,3]
plt.fill_between(z,m)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.hexbin(z,m,gridsize=(30,30))
plt.show()
```

```python
plt.plot(z,m)
plt.figure(figsize=(100,200))
plt.show()
```
